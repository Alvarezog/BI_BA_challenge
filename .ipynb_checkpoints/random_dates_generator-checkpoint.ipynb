```python
import pandas as pd
import numpy as np
import datetime
import random
```

```python
data = pd.read_csv('businessBIBA_challenge.csv')
data.head()
```

```python
#Function that generates random dates from start date to the actual date.
def random_date(start,end):
    new_time = start + datetime.timedelta(seconds= int((end-start).total_seconds()*random.random()))
    return new_time
```

```python
#Generate a list of coherent random dates, where the dates follow the response, risk, delivered/approved 
#and transaction chronological order. Time limit for risk response 

for i in range(len(data)):
    if i==0:
        data.at[i,"UPDATE"]= random_date(datetime.datetime(2021,1,1,0,0,1),datetime.datetime.today())
        
    elif data.loc[i]['ID'] != data.loc[i-1]['ID']:
        data.at[i,"UPDATE"]= random_date(datetime.datetime(2021,1,1,0,0,1),datetime.datetime.today())
    
    #Time limit for risk response is 7 days.
    elif data.loc[i]['ID'] == data.loc[i-1]['ID'] and data.loc[i]['STATUS']=='RISK':
        start = data.loc[i-1]['UPDATE']
        end = start + datetime.timedelta(days= 7)
        data.at[i,"UPDATE"]= random_date(start,end)
    
    #Time limit for approved or rejected response is 7 days.
    elif data.loc[i]['ID'] == data.loc[i-1]['ID'] and data.loc[i]['STATUS']=='APPROVED':
        start = data.loc[i-1]['UPDATE']
        end = start + datetime.timedelta(days= 7)
        data.at[i,"UPDATE"]= random_date(start,end)
    elif data.loc[i]['ID'] == data.loc[i-1]['ID'] and data.loc[i]['STATUS']=='REJECTED':
        start = data.loc[i-1]['UPDATE']
        end = start + datetime.timedelta(days= 7)
        data.at[i,"UPDATE"]= random_date(start,end)
    
    #Time limit for delivered is 15 days.
    elif data.loc[i]['ID'] == data.loc[i-1]['ID'] and data.loc[i]['STATUS']=='DELIVERED':
        start = data.loc[i-1]['UPDATE']
        end = start + datetime.timedelta(days= 15)
        data.at[i,"UPDATE"]= random_date(start,end)
    
    #No time limit for transactions.
    elif data.loc[i]['ID'] == data.loc[i-1]['ID'] and pd.isnull(data.loc[i]['STATUS']):
        start = data.loc[i-1]['UPDATE']
        end = datetime.datetime.today()
        data.at[i,"UPDATE"]= random_date(start,end)
```

```python
data.to_csv('businessBIBA_challenge_dates.csv',index=False,na_rep='NULL')
```
